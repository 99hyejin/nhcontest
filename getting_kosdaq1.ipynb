```python
import os
import pandas as pd
import csv
```

```python
folder_path = 'C:/NH/KOSDAQ/'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
```

```python
for file in csv_files:
    date, _ = os.path.splitext(file)  # Extract the date from the file name
    file_path = os.path.join(folder_path, file)  # Create the full file path
    df = pd.read_csv(file_path, quoting=csv.QUOTE_NONE, encoding='cp949')
    # Add a 'date' column with the extracted date
    df['date'] = date
    
    # Append the data to the merged DataFrame
    kqut = kqut.append(df, ignore_index=True)

# Save the merged DataFrame to a new CSV file if needed
kqut.to_csv('merged_data.csv', index=False, encoding='utf-8')
```

```python
folders = os.listdir('C:/NH/KOSDAQ/')
kqut = pd.DataFrame()
```

```python
for i in range(len(folders)):
    date, k = folders[i].split('.')
    if k == 'csv':
        file = 'C:/NH/KOSDAQ/'+folders[i]
        df = pd.read_csv(file, quoting=csv.QUOTE_NONE, encoding='cp949')
        df = df[df['시장구분']=='KOSDAQ']
        df['date'] = date
        kqut = pd.concat([kqut, df])
```

```python
kqut.to_csv('C:/NH/KOSDAQ_QUT.csv',index=False, encoding = 'euc-kr')
```
